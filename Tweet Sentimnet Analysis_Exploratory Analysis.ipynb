## 2.1 Label distribution plot

```python
import matplotlib.pyplot as plt
class_each = df['label_name'].value_counts()
plt.figure(figsize=(8, 5))
class_each.plot(kind='bar')
plt.title('Sentiment Labels Distribution')
plt.xlabel('Sentiment labels')
plt.ylabel('# tweets')
plt.show()
```

## 2.2. EDA – Tweet Word Counts Distribution 

```python
df['character'] = df['text'].astype(str).str.len()
df['word_count'] = df['text'].astype(str).str.split().str.len()
plt.figure(figsize=(8, 5))
plt.hist(df['word_count'], bins=25)
plt.title('Tweet Word Counts Distribution')
plt.xlabel('# words')
plt.ylabel('# tweets')
plt.show()
```

## 2.3 VADER Compound Score Distribution
Overlap between classes suggests ambiguity and explains why classification is difficult.

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
tweetana = SentimentIntensityAnalyzer()
def calcvader(text):
    scores = tweetana.polarity_scores(str(text))
    return pd.Series(scores)
vscores = df['text'].apply(calcvader)
vdf = pd.concat([df.copy(), vscores], axis=1)
```

```python
plt.figure(figsize=(8, 5))
for three_label in ['negative', 'neutral', 'positive']:
    subset = vdf[vdf['label_name'] == three_label]
    plt.hist(subset['compound'], bins=15, alpha=0.25, label= three_label)
plt.title('VADER Compound Score of Three Classes')
plt.xlabel('VADER Compound Scores')
plt.ylabel('# tweets')
plt.legend()
plt.show()
```
